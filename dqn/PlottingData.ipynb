```python
import pandas as pd
import json
import matplotlib.pyplot as plt
```

```python
with open('./data_dqn5/DataValue3.json') as f:
    data = json.load(f)
```

```python
df = pd.DataFrame(data)
```

```python
run = []
steps = []
for t in df:
    run.append(df[t]['run'])
    steps.append(df[t]['steps'])
```

```python
x = run
y = steps
```

```python
plt.plot(x, y, color='gray')
plt.ylabel('steps')
plt.xlabel('trial')
plt.title('DQN 3x3')
plt.savefig('4MissingDQN3x3.png', transparent=True)
```
